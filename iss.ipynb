```python
import requests
import matplotlib.pyplot as plt
```

```python
import time

x = []
y1 = []
y2 = []

for _ in range(100):
    response = requests.get("https://api.wheretheiss.at/v1/satellites/25544").json()
    y1.append(response["latitude"])
    y2.append(response["longitude"])
    x.append(response["timestamp"])
    time.sleep(10)
```

```python
plt.plot(x, y1, label="Latitude", color="red")
plt.plot(x, y2, label="Longitude", color="blue")
plt.xlabel("Timestamp")
plt.ylabel("Location")
plt.title("Track the ISS")
plt.legend()
plt.grid(True)
plt.show()
```
